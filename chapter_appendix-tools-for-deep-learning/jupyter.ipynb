```python
import requests
import pandas as pd
import time
```

```python
API_KEY = "8265bd1679663a7ea12ac168da84d2e8"
BASE_URL = "https://api.themoviedb.org/3/movie/top_rated"

movies_data = []
```

```python
for page in range(1, 472):
    params = {
        "api_key": API_KEY,
        "language": "en-US",
        "page": page
    }

    response = requests.get(BASE_URL, params=params)
    data = response.json()

    for movie in data.get("results", []):
        movies_data.append({
            "movie_id": movie.get("id"),
            "title": movie.get("title"),
            "overview": movie.get("overview")
        })
    # print(f"Done for page {page}")
    time.sleep(0.25)  # prevents rate limiting
```

```python
df = pd.DataFrame(movies_data)

# df = df.dropna(subset=["overview"])

df.head()
```

```python
df["overview"] = df["overview"].str.lower()
df.head()
```

```python
# Remove html tags
import re

df["overview"] = df["overview"].apply(
    lambda x: re.sub(r"<.*?>", "", x)
)
df.head()
```

```python
# Remove urls
df["overview"] = df["overview"].apply(
    lambda x: re.sub(r"http\S+|www\S+|\S+@\S+", "", x)
)
df.head()
```

```python
# Remove extra white spaces
df["overview"] = df["overview"].apply(
    lambda x: re.sub(r"\s+", " ", x).strip()
)
```

```python
#Remove punctuation and number
df["overview"] = df["overview"].apply(
    lambda x: re.sub(r"[^a-z\s]", "", x)
)
```

```python
from textblob import TextBlob
df["overview"] = df["overview"].apply(
    lambda x: TextBlob(x).correct().string
)
```

```python
df["tokens"] = df["overview"].apply(lambda x: x.split())

# Removing stop words
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

df["tokens"] = df["tokens"].apply(
    lambda tokens: [w for w in tokens if w not in stop_words]
)
```

```python
df.head()
```

```python
# Lemmatization
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

df["tokens"] = df["tokens"].apply(
    lambda tokens: [lemmatizer.lemmatize(w) for w in tokens]
)
```
