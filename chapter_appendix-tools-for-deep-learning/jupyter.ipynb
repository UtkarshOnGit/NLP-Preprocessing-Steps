```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")
```

```python
doc.text
```

```python
doc[0]
```

```python
# Course Grain Part Of Speech
doc[0].pos_
```

```python
# Fine Grain Part Of Speech
doc[0].tag_
```

```python
spacy.explain('NNP')
```

```python
doc2 = nlp("India is going to win the World Cup")

for a in doc2:
  print(f"{a.text}----> , {a.pos_} , {a.tag_} , {spacy.explain(a.tag_)}")
```
